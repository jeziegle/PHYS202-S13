```python
theta = arange(-pi,pi,.1)
y = (cos(2*theta)*cos(theta/2))**2
plot(theta,y, 'r-')
plot(theta,y, 'b*')
xlabel(r"$\theta$ (rad)")
ylabel("intensity")
xticks(arange(-pi,5*pi/4,pi/4),('$-\pi$','-$3\pi/2$','-$\pi/2$','-$\pi/4$',0,'$\pi/4$','-$\pi/2$','$3\pi/4$','$\pi$'))
show()
```

```python
t = arange(-4.5,4.6,.1)
f = (t+1)/(t-1)
g = exp(-t)

plot(t, f,label = "(t+1)/(t-1)")
plot(t, g,'r:',label = "exp(-t)")
legend(loc='lower left')
xlabel("time (seconds)")
ylabel("f(t), g(t)")
ylim([-10,10])
yticks(arange(-10,15,5))
annotate("discontinuous\nhere", xy = (1.2,-2.5))
show()
```

```python
def is_anagram(s1,s2):
    if len(s1)!=len(s2):
        return False
    for letter in s1:
        if not letter in s2:
            return False
    return True
```
