```python
x=arange(-pi,pi,.1)
s=-(1/cosh(x))**2
i=1j
p_0=tanh(x)
p_half=conj((i*.5-tanh(x))/(i*.5+1)*exp(i*.5*x))
p_1=conj((i-tanh(x))/(i+1)*exp(i*x))
p_2=(i*2.0-tanh(x))/(i*2.0+1)*exp(i*2.0*x)
plot(x,s,label="$u(x)$")
plot(x,p_0**2,color="red",label="$\psi^2,k=0.0$")
plot(x,p_half.conjugate()*p_half,color="red",linestyle='--',label="$\psi^2,k=.5$")
plot(x,p_1.conjugate()*p_1,color="red",linestyle='-.',label="$\psi^2,k=1.0$")
plot(x,p_2.conjugate()*p_2,color="red",linestyle=':',label="$\psi^2,k=2.0$")
ylim((-1.3,1.3))
legend(loc="lower left",labelspacing=.2,prop={'size':8})
xlabel("$x(rad)$")
ylabel("$U(x),\psi^2(x)$")
xticks(arange(-pi,1.5*pi,pi/2),('$-\pi$','-$\pi/2$',0,'$\pi/2$','$\pi$'))
yticks(arange(-1.0,1.5,.5))
grid(True)
show()
```

```python
sunspotdata = loadtxt("sunspots.txt",float)
year = (sunspotdata[:1001,0]+1)/12+1749
number = sunspotdata[:1001,1]
numave=[]
for i in range(1001):
    numave.append(sum(number[:i])/i)
plot(year, number, color = 'red',linewidth = .5,label = "$Monthly$ $Sunpots$ $Since$ $1749$")
plot(year, numave, color = 'purple',linewidth = 1.5,label = "$Average$ $of$ $Monthly$ $Sunspots$")
xlabel("$Time$ $(Year A.D.)$")
ylabel("$Sunpots$ $per$ $Month$")
xticks(arange(1750,1840,15))
yticks(arange(0,300,50))
grid(True)
legend()
show()
```

```python
stardata = loadtxt("stars.txt",float)
temp = stardata[:,0]
mag = stardata[:,1]
scatter(temp, mag, color = 'red', s=.2, label = "$Star$ $Temp.$ $vs.$ $Stellar$ $Magnitude$")
xlabel("$Temp.$ $(K)$")
ylabel("$Magnitude$")
xticks(arange(0,16000,2000))
xlim([16000,0])
ylim([20,-5])
annotate('$Main$ $Sequence$',xy = (5000, 5),xytext = (4000,4),arrowprops = dict(facecolor = 'black', shrink = .2,frac = .5),)
annotate('$White$ $Dwarfs$',xy = (7500,14),xytext = (9500,17),arrowprops = dict(facecolor = 'black', shrink = -.1,frac = .3))
annotate('$Horizontal$ $Branch$',xy = (5000,0),xytext = (5500,-2.5),arrowprops = dict(facecolor = 'black', shrink=.22, frac=.4))
legend(loc='lower left',prop={'size':11})
show()
```
