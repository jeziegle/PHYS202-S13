Exercise 1: HitOrMiss MC Integration
-

```python
f = lambda x: x**(.5) + sin(x)
x = arange(0.,15.,0.001)
myplot = plot(x,f(x), lw = 3)
ylabel(r'$x^{1/2} + \sin(x)$', fontsize = 20)
xlabel(r'$x$', fontsize = 20)
xlim(0.,15.)
show()
```

```python
xmin = 1.7
xmax = 12.4

subx = x[logical_and((x > xmin),(x < xmax))]
ymax = max(f(subx))
ymin = 0.
myplot = plot(x, f(x), lw = 3)
xlim(0., 15.)

plot([xmin,xmin], [ymin, ymax], color = 'k', linestyle = '--')
plot([xmax,xmax], [ymin, ymax], color = 'k', linestyle = '--')
plot([xmin,xmax], [ymax, ymax], color = 'k', linestyle = '--')

show()
```

```python
N = 10**7
samples_x = xmin + (xmax-xmin)*random_sample(N)
samples_y = ymin + (ymax-ymin)*random_sample(N)

#myplot = plot(x,f(x),lw=3)
xlim(0.,15.)

#plot(samples_x[1:N],samples_y[1:N],'.')

#plot([xmin,xmin], [ymin, ymax], color = 'k', linestyle = '--')
#plot([xmax,xmax], [ymin, ymax], color = 'k', linestyle = '--')
#plot([xmin,xmax], [ymax, ymax], color = 'k', linestyle = '--')

#show()
```

```python
newmask = (samples_y < f(samples_x))
sum(newmask)
```

```python
I = (xmax-xmin)*(ymax-ymin)*sum(newmask)/newmask.size

Iana = lambda x: (2./3.)*x**(3/2.)-cos(x)
Iexact = Iana(xmax)-Iana(xmin)
Ierr = abs(100*(Iexact-I)/Iexact)
print "Iexact: ",Iexact
print "Ihitormiss: ",I
print "Ierror: ",Ierr,"%"
#this integration method won't easily go to .001% error
```

Sample Mean Method
-

```python
N = 10000
ave = 0
interval = float((xmax-xmin)/N)

for i in range(N+1):
    ave += f(xmin+i*interval)/N

Imean = ave * (xmax-xmin)

plot([0,15], [ave,ave])
plot(x, f(x))
xlim(0.,15.)
show()

print "Sample Mean: ", Imean
print "Exact: ", Iexact
print "Percent Error: ", abs(Imean - Iexact)/Iexact
#The mean method is a lot a lot better than MC's method
```
