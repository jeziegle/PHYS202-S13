Using MatPlotLib to make pretty graphs

```python
#simple line plot
t = arange(0.0,1.01,.01)
c = cos(4*pi*t)
s = sin(4*pi*t)
```

```python
plot(t, s, linewidth=2.5,label="$sin(4{\pi}t)$")
plot(t, c, color="red", linewidth=2.5, label="$cos(4{\pi}t)$")
xlabel("Time (s)")
ylabel("Voltage (volts)")
title("Sine plot",fontsize = 20)
legend()
grid(True)
savefig("plot.pdf")
show()
```
