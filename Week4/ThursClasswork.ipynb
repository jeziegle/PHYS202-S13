```python
X, Y = meshgrid(arange(-3,3,.06),arange(-3,3,.06))
V = dipolePotential(X, Y, 3*10**-9, 1.) + pointPotential(X, Y, 1*10**-9, 0, .5)

from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = fig.gca(projection = '3d')
p = ax.plot_surface(X, Y , V ,rstride = 1, cstride = 1, linewidth = 0, cmap = cm.flag)
show()
levels = arange(-50,50,5)
contour(X, Y, V, levels, linewidths = 4, cmap = cm.Accent_r)
show()
```
