Boolean Masks
-

```python
theta = arange(0., 2*pi, pi/100.)
plot(theta,sin(theta))
show()
```

```python
mask = (sin(theta)>0) & (sin(theta)<0.5)

plot(theta, sin(theta), theta[mask], sin(theta[mask]),)
show()
```

```python
y = sin(theta[mask])
jj = (y == max(y))
y[jj] = NaN
plot(theta, sin(theta))
plot(theta[mask], y,linewidth = 4)
show()
```

Interpolation
-

```python
theta = arange(0., 2*pi+.01, pi/10)
mask = (theta == 0.5)
```

```python
newTheta = 0.5; newSin = 0.
from scipy.interpolate import interp1d
#help(interp1d)
sinInterp = interp1d(theta, sin(theta))
newSin = sinInterp(newTheta)
print newSin
plot(theta, sin(theta), newTheta, newSin, 'r+', markersize = 20, markeredgewidth = 3)
show()
```

```python
thetaNew = arange(0, 2*pi+.01, pi/30)
sinNew = interp1d(theta, sin(theta))(thetaNew)
plot(theta, sin(theta), thetaNew, sinNew, 'r.')
show()
```

```python
plot(thetaNew, sinNew-sin(thetaNew))
ylabel("difference between interp and exact values")
show()
```

```python
sinNewC = interp1d(theta, sin(theta), kind = 'cubic')(thetaNew)
subplot(2, 1, 1)
plot(thetaNew, sinNew - sin(thetaNew), thetaNew, sinNewC - sin(thetaNew), 'r')
subplot(2, 1, 2,)
plot(thetaNew, sinNewC - sin(thetaNew), 'r')
show()
```

Numerical Differentiation
-

```python
x = arange(0., 10.1, .1)
y = x**5
```

```python
dy = diff(y) #y(n+1) - y(n)
#help(diff)
dx = diff(x)
dydx = dy/dx
print len(dydx)
```

```python
dydx1 = zeros(y.shape,float)
dydx1[:-1] = diff(y)/diff(x)
dydx1[-1] = (y[-1] - y[-2])/(x[-1] - x[-2])
```

$y = x^5$, $\frac{dy}{dx} = 5 x^4$

```python
dydxExact = 5*x**4
dydxCenter = zeros(y.shape,float)
dydxCenter[1:-1] = (y[2:]-y[:-2])/(x[2:]-x[:-2])
dydxCenter[0] = (y[1]-y[0])/(x[1]-x[0])
dydxCenter[-1] = (y[-1]-y[-2])/(x[-1]-x[-2])
```

```python
plot(x, dydx1, label = 'Forward')
plot(x, dydxExact, label = 'Exact')
plot(x, dydxCenter, label = 'Centered')
legend(loc = 'upper left')
show()
```

```python
percerr1 = 100*abs(dydx1[1:] - dydxExact[1:])/dydxExact[1:]
percerr2 = 100*abs(dydxCenter[1:] - dydxExact[1:])/dydxExact[1:]
semilogy(x[1:], percerr1, label = 'Percent error: Forward')
semilogy(x[1:], percerr2, label = 'Percent error: Centered')
legend(loc = 'upper right')
show()
```
