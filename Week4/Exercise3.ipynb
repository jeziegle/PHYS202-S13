```python
def pointPotential(x , y, q, posx, posy):
    #Finds potential at x and y due to a point charge of charge q at posx, and posy
    k = 8.987551787368176*10**9;
    return (k*q/((x+posx)**2+(y+posy)**2)**.5)

def dipolePotential(x, y, q, d, theta):
    return pointPotential(x, y, q, x-d*cos(theta), y-d*sin(theta)) + pointPotential(x, y, -q, x+d*cos(theta), y+d*sin(theta))
```

```python
dx = .06
dy = .06
X, Y = meshgrid(arange(-2.,2.,dx),arange(-2.,2.,dy))
V = dipolePotential(X, Y, 1/(9.*10.**9.), 1, 0)

from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = fig.gca(projection = '3d')
p = ax.plot_surface(X, Y, V, rstride = 1, cstride = 1, linewidth = 0, cmap = cm.flag)
cb = fig.colorbar(p, shrink = .7)
```

```python
dx = .06
dy = .06
X, Y = meshgrid(arange(-4.,4.,dx),arange(-4.,4.,dy))
V = pointPotential(X, Y, 1*10**-9, 0, -2) + pointPotential(X, Y, 1*10**-9, 0, 2) + pointPotential(X, Y, -1*10**-9, 2, 0) + pointPotential(X, Y, -1*10**-9, -2, 0)

from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = fig.gca(projection = '3d')
p = ax.plot_surface(X, Y, V, rstride = 1, cstride = 1, linewidth = 0, cmap = cm.prism)
cb = fig.colorbar(p, shrink = .7)
```
