3-D Plotting and Functions
=

$V(x,\hspace{1 mm}y) = \frac{kq}{\sqrt{x^2\hspace{1 mm}-\hspace{1 mm}(y\hspace{1 mm}-\hspace{1 mm}d)^2}}-\frac{kq}{\sqrt{x^2\hspace{1 mm}-\hspace{1 mm}(y\hspace{1 mm}+\hspace{1 mm}d)^2}}$

```python
#make the x-y grid with step size .06 to cover the plane -2 < x < 2 and -2 < y < 2
dx = 0.06
dy = 0.06
x = arange(-2., 2., dx)
y = arange(-2., 2., dy)
X, Y = meshgrid(x,y)
#make the function, V(x, y)
k = 1; q = 1; d = 1;
V = (k*q/(X**2+(Y-d)**2)**.5)-(k*q/(X**2+(Y+d)**2)**.5)

from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = fig.gca(projection = '3d')
p = ax.plot_surface(X, Y, V, rstride = 1, cstride = 1, linewidth = 0, cmap = cm.flag)
cb = fig.colorbar(p, shrink = .7)
```

```python
#represent 3d data by countours
levels = arange(-10,10,.7)

contour(X, Y, V, levels, linewidths = 4, cmap = cm.Accent_r)
show()
```

Ex. 1: Make a 3-D plot of the function $G(x, y) = e^{(-x^2-y^2)}$
-
over the range -6< (x,y) <6 using 51 bins in $x$ and 51 bins in $y$
-

```python
dx = 12./51.
dy = 12./51.
x = arange(-6,6,dx)
y = arange(-6,6,dy)
X, Y = meshgrid(x, y)
G = exp(-X**2-Y**2)

from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = fig.gca(projection = '3d')
p = ax.plot_surface(X, Y, G, rstride = 1, cstride = 1, cmap = cm.rainbow, linewidth = 0)
cb = fig.colorbar(p, shrink = .7)
```

Ex 2: Determine the electric potential for a quadrupole consisting of q = -1.0 nC at both(x = -2 and +2, y = 0) and q = 1.0 nC at both (x = 0 and y = -2 and +2). Make a surface plot and a contour plot on the range -4<x, y<4 using 101 bins in x and 101 in y. What is the value of the function at V(20,40)?
-

```python
#make the x-y grid with step size .06 to cover the plane -2 < x < 2 and -2 < y < 2
dx = 8./101
dy = 8./101
x = arange(-4., 4., dx)
y = arange(-4., 4., dy)
X, Y = meshgrid(x,y)
#make the function, V(x, y)
#k is in N*m^2*C^-2
k = 8.987551787368176*10**9; q = 1*10**-9; d = 2;
V = (k*q/(X**2+(Y-d)**2)**.5)+(k*q/(X**2+(Y+d)**2)**.5)-(k*q/((X+d)**2+Y**2)**.5)-(k*q/((X-d)**2+Y**2)**.5)

from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = fig.gca(projection = '3d')
p = ax.plot_surface(X, Y, V, rstride = 1, cstride = 1, linewidth = 0, cmap = cm.prism)
cb = fig.colorbar(p, shrink = .7)
```

```python
levels = arange(-15.,15.,2)
contour(X, Y, V, levels, linewidths = 4, cmap = cm.rainbow)
show()
```
