Q2 - Weibull Distribution
-

```python
alpha = 1.5
beta = 6
f = lambda x: alpha*beta**-alpha*x**(alpha-1)*exp(-(x/beta)**alpha)
F = lambda x: exp(-(x/beta)**alpha)
Finv = lambda z: beta*(-log(z))**(1/alpha)
```

```python
import numpy.random
N = 10e4
z = numpy.random.random_sample(N)
x = arange(0,25.1,.1)

hist(Finv(z),50,normed=True,label="Sampled Distribution")
plot(x,f(x),'r-',linewidth=2.5,label="True Distribution")
legend(loc='upper right')
show()
```

Q3 - Beta(4,3) distibution
-

```python
def distribute1D(x,prob,N):
    """Takes any numerical distribution prob, on the interval defined by the array x, and returns an array of N sampled values that are statistically the same as the input data."""
    Fx = cumsum(prob)/sum(prob)
    z = numpy.random.random_sample(N)
    Finv = x[searchsorted(Fx,z)]
    
    return Fx,Finv
```

```python
N = 10.e4
x = arange(0.,1.01,.01)
f = lambda x: 60*x**3*(1-x)**2
Fx,Finv = distribute1D(x,f(x),N)
hist(Finv,50,normed=True,label="Sampled Distrib.")
plot(x,f(x),'r-',linewidth=2.,label="True Distrib.")
plot(x,Fx,'k-',linewidth=2.,label="CDF")
legend(loc='upper left')
show()
```
