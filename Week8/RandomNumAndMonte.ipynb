```python
x = arange(0,pi+.01,.01)
f = lambda x: sin(x)/2.
F = lambda x: (cos(0)-cos(x))/2.

plot(x,f(x),label="True distribution",lw=3)
plot(x,F(x),label="Cumulative function",lw=3)

xlim(0.,3.5)
ylim(0.,1.1)

legend(loc=0)
show()
```

```python
N = 10000

z = numpy.random.random_sample(N);

Finv = lambda z:arccos(cos(0)-2*z)
    
xsamples = Finv(z);

hist(xsamples,50,normed=True,label="Sampled distribution")
plot(x,f(x),label="True distribution",lw=3,color="red")

xlim(0.,pi)
ylim(0.,0.7)

legend(loc=0)
show()
```

```python
lamb = 1
#integral of c*e^(-c*x) is -e^(-c*x)
f = lambda x: lamb*exp(-lamb*x)
F = lambda x: 1-exp(-lamb*x)
x = arange(0.,5.001,.001)    
plot(x,F(x))
plot(x,f(x))
show()
```

```python

Finv = lambda z: -log(1-z)/lamb
    
z = numpy.random.random_sample(100000)

hist(Finv(z),50,normed=True,label="CDF")
plot(x,f(x),'r-',label="Actual Function")
legend()
show()
```

```python
lamb = 1
#integral of c*e^(-c*x) is -e^(-c*x)
f = lambda x: lamb*exp(-lamb*x)
x = arange(0.,5.001,.001)
dist = f(x)

Fx = cumsum(dist)/sum(dist)
z = numpy.random.random_sample(10000)
Fxinv = x[searchsorted(Fx,z)]
hist(Fxinv,50,normed=True)
plot(x,f(x))
show()
```

```python
def distribute1D(x,prob,N):
    """Takes any numerical distribution prob, on the interval defined by the array x, and returns an array of N sampled values that are statistically the same as the input data."""
    Fx = cumsum(prob)/sum(prob)
    z = numpy.random.random_sample(N)
    xsamples = x[searchsorted(Fx,z)]
    
    return xsamples
```

```python
import numpy.random
x = arange(-.2,.201,.001)
a = 10e-6
d = 15e-6
L = 1
lamb = 632.8e-9
intens = lambda x: (sinc(a*x/(lamb*L))**2)*(cos(2*pi*d*x/(lamb*L))**2)
Fxinv = distribute1D(x,intens(x),10000)
n,bins,patches=hist(Fxinv,50,normed=True,alpha=1.0)
plot(x,max(n)*intens(x),'r-')
xlabel("x (m)")
ylabel("intensity")
show()
```
