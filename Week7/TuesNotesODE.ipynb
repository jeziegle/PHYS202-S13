```python
g = 9.8
l = 0.5

theta_app = lambda t: th0*cos(sqrt(g/l)*t)
from scipy.integrate import odeint
```

```python
def deriv(theta,t):
    g = 9.8
    l = .5
    return array([theta[1],-(g/l)*sin(theta[0])])
```

```python
t = arange(0.,10.01,.01)
th0 = 20.*pi/180
omega0 = 0
theta_init = array([th0,omega0])
theta = odeint(deriv,theta_init,t)
```

```python
plot(t,theta_app(t),'b-',t,theta[:,0],'r--')
show()
```

```python
th0 = 50.*pi/180

theta_init = array([th0,omega0])
theta = odeint(deriv,theta_init,t)

plot(t,theta_app(t),'b-',t,theta[:,0],'r--')
show()
```

```python
theta_init1 = array([0,omega0])
theta1 = odeint(deriv,theta_init1,t)
theta_app1 = lambda t: 0*cos(sqrt(g/l)*t)
plot(t,theta_app1(t),'b-',t,theta1[:,0],'r--')

theta_init2 = array([pi/4.,omega0])
theta2 = odeint(deriv,theta_init2,t)
theta_app2 = lambda t: pi/4.*cos(sqrt(g/l)*t)
plot(t,theta_app2(t),'g-',t,theta2[:,0],'m--')

theta_init3 = array([pi/2,omega0])
theta3 = odeint(deriv,theta_init3,t)
theta_app3 = lambda t: pi/2*cos(sqrt(g/l)*t)
plot(t,theta_app3(t),'b-',t,theta3[:,0],'r--')

theta_init4 = array([3*pi/4,omega0])
theta4 = odeint(deriv,theta_init4,t)
theta_app4 = lambda t: 3*pi/4*cos(sqrt(g/l)*t)
plot(t,theta_app4(t),'g-',t,theta4[:,0],'m--')

theta_init5 = array([pi,omega0])
theta5 = odeint(deriv,theta_init5,t)
theta_app5 = lambda t: pi*cos(sqrt(g/l)*t)
plot(t,theta_app5(t),'b-',t,theta5[:,0],'r--')

show()
```

The pendulum's period increases as $\theta_0$ increases. The trajectory at 0 is zero and the trajectory at $\pi$ is really long
