```python
dydxfunc = lambda x,y: x+2*y
dx = .25
x1 = arange(0.,1.+dx,dx)
dydx = zeros(x1.shape,float)
y1 = zeros(x1.shape,float)
for i in range(1,len(dydx)):
    dydx[i] = dydx[i-1]+(1+2*dydxfunc(x1[i-1],y1[i-1]))*dx
    y1[i] = y1[i-1]+dydx[i-1]*dx

print y1[1/dx]
```

```python
dx = .02
x2 = arange(0.,1.+dx,dx)
dydx = zeros(x2.shape,float)
y2 = zeros(x2.shape,float)
for i in range(1,len(dydx)):
    dydx[i] = dydx[i-1]+(1+2*dydxfunc(x2[i-1],y2[i-1]))*dx
    y2[i] = y2[i-1]+dydx[i-1]*dx

print y2[1/dx]
```

```python
x = arange(0,1.01,.01)
yexact = .25*exp(2*x)-.5*x-.25
plot(x1,y1,label="dx = .25")
plot(x2,y2,label="dx = .02")
plot(x,yexact,label="Exact")
legend(loc='upper left')
show()
#The smaller step size gives a better result, as expected
```

```python
dx = .02
x2 = arange(0.,1.+dx,dx)
dydx = zeros(x2.shape,float)
y2 = zeros(x2.shape,float)
for i in range(1,len(dydx)):
    dydx[i] = dydx[i-1]+(1+2*dydxfunc(x2[i-1],y2[i-1]))*dx
    y2[i] = y2[i-1]+dydx[i]*dx

print y2[1/dx]
```

```python
x = arange(0,1.01,.01)
yexact = lambda x: .25*exp(2*x)-.5*x-.25
plot(x1,y1,label="dx = .25")
plot(x2,y2,label="dx = .02")
plot(x,yexact(x),label="Exact")
legend(loc='upper left')
show()
plot(x1,y1-yexact(x1),label="dx=.25 difference")
plot(x2,y2-yexact(x2),label="dx=.02 difference")
legend(loc='lower left')
show()
```
