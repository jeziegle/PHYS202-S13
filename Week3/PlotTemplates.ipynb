```python
#Create numpy array of y values for plugging into equations
y = arange(-5, 5, .1)
#Create array of cosh(y) with one value for each value in y
E = cosh(y)
#Create array of sinh(y) with one value for each value in y
absp = sinh(y)
```

Line Plots
-

```python
#Plot E with respect to y, setting linewidth, label, and color to something nice looking
plot(y,E, linewidth = 1.5, color = 'purple', label = "$Plot$ $Number$ $1$")
plot(y,absp, linewidth = 1.5, color = 'red', label = "$Plot$ $Number$ $2$")
#Show legend and set location to make all data visible
legend(loc = 'lower right')
#Remove boxing by setting top and right "spines" to colorless (not there)
ax = gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
#Center axes at zero
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
#Label x and y axes
xlabel("$X$ $axis$")
ylabel("$Y$ $axis$")
#Turn grid on
grid(True)
show()
```

Subplots
-

```python
#Two rows and one column of plots, first plot
subplot(211)
#Do same stuff as above to make plots nice
plot(y,E, linewidth = 1.5, color = 'purple', label = "$Plot$ $Number$ $1$")
legend(loc = 'best')
ylabel("$Y$ $axis$")
ax = gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')

#Two rows and one column of plots, second plot
subplot(212)
#Do same stuff as above to make plots nice
plot(y,absp, linewidth = 1.5, color = 'red', label = "$Plot$ $Number$ $2$")
legend(loc = 'lower right')
xlabel("$X$ $axis$")
ylabel("$Y$ $axis$")
ax = gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
show()
```
