```python
f = lambda x: exp(-x**2)
x = arange(0,3.1,.1)

import scipy.integrate as integ 

eSimps = zeros(shape(x),float)
for i in range(1, len(x)):
    eSimps[i] = integ.simps(x[:i],f(x[:i]))
    
print eSimps
plot(x, eSimps, label = "E(x)")
legend(loc = 'upper right')
xlabel("X")
ylabel("E(x)")
show()
```

```python
def H(n,x):
    Hiless1 = 0*x
    Hi = 1+0*x
    for i in range(0, n):
        Himore1 = 2.*x*Hi-2.*i*Hiless1
        Hiless1 = Hi
        Hi = Himore1
        
    Her = Hi
    return Her
```

```python
from math import factorial
psi = lambda n, x : exp(-x**2./2.)/sqrt((2.**n)*factorial(n)*sqrt(pi)) * H(n,x)


x = arange(-4, 4, .01)
plot(x,psi(0,x),label="n = 0")
plot(x,psi(1,x),label="n = 1")
plot(x,psi(2,x),label="n = 2")
plot(x,psi(3,x),label="n = 3")
legend(loc = 'lower right')
xlabel("x")
ylabel("Amplitude")
show()
```
